```python
def visualize_metrics(metrics_data):
    df = pd.DataFrame(metrics_data)
    
    # Plot accuracy
    plt.figure(figsize=(8, 6))
    plt.bar(df.index, df['accuracy'])
    plt.xlabel("Model")
    plt.ylabel("Accuracy")
    plt.title("Model Accuracy Comparison")
    plt.show()

    # Plot F1-scores per class
    plt.figure(figsize=(10, 6))
    categories = [col for col in df.columns if col.endswith("_f1-score")]
    for cat in categories:
        plt.plot(df.index, df[cat], label=cat.split("_")[0])
    plt.xlabel("Model")
    plt.ylabel("F1-Score")
    plt.title("Per-Class F1-Score Comparison")
    plt.legend()
    plt.show()

visualize_metrics(metrics_data)
```
